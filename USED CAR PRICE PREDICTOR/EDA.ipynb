# After data cleaning now I decided to use Linear Regression for the prediction.

I have to make an assumption and make them true so it can give me support for the use the Linear Regression. Because I know that Linear Regression algorithm works only when the relation with features with the target column is in a linear manner.
I don't need to focus the categorical columns because those will be converted as dummies variable.(One- hot encoding)
all I have problem with numerical columns. like Mileage, EngineV, and Year

```python
# importing libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set() # all plot will cover on seaborn style
```

```python
# loading the data
car = pd.read_csv('Car_cleaned.csv')
car.head()
```

```python
# plotting all the three numerical columns
plt.figure(figsize=(16,4))

plt.subplot(1,3,1)
sns.scatterplot(x=car['Year'], y=car['Price']);
plt.title('Year Vs Price');

plt.subplot(1,3,2)
sns.scatterplot(x=car['EngineV'], y=car['Price']);
plt.title('EngineV Vs Price');

plt.subplot(1,3,3)
sns.scatterplot(x=car['Mileage'], y=car['Price']);
plt.title('Mileage Vs Price');
```

# The relation is not linear therefore I can not run Linear Regression in here.
We have to transform these column to run linear regression.

```python
## Checking the price
sns.distplot(car['Price']);
plt.title('Price')
```

The problem lies with the price, the price we have in the data is not linear.
Common approach to transform the data is log transformation.
Numpy has a method which will calculate natural log which is np.log(x) it return natural logarithm of a number or arrays.

```python
# transform the price
log_price = np.log(car['Price'])

# adding another column 
car['log_price'] = log_price

# checking the data again
car.head()
```

```python
# plotting all the three numerical columns again with log price
plt.figure(figsize=(16,4))

plt.subplot(1,3,1)
sns.scatterplot(x=car['Year'], y=car['log_price']);
plt.title('Year Vs log_price');

plt.subplot(1,3,2)
sns.scatterplot(x=car['EngineV'], y=car['log_price']);
plt.title('EngineV Vs log_price');

plt.subplot(1,3,3)
sns.scatterplot(x=car['Mileage'], y=car['log_price']);
plt.title('Mileage Vs log_price');
```

# Now the whole picture has changed.
Here we can observe the linearity of the data with Price column

```python
# droping the Original price column that is no longer needed
car.drop('Price', axis=1, inplace=True)
```

# Now check the Multicollinearity.
this occurs when the feature of the data is corealeted with other feature.

```python
car.columns
```

There can be some multicollinearity in the data and can be assumed,like the newer the car will have its mileage lower. so Year and Milage will be correleated.
I will use the VIF or Variance inflation factor to check the multicolinearity.
The VIF produce a measure which estimate how much larger the square root of the standerd error of the estimate is compared to a situation wher the variable was completely uncorreleated with other features.

```python
# import VIF from statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

# taking only numerical column beacuse the categorical will be converted as dummy variable
var = car[['Mileage', 'Year', 'EngineV']]
# var become a dataframe with only three columns


# creating a dataframe
vif_df = pd.DataFrame()

# adding column
vif_df['VIF'] = [vif(var.values, i) for i in range(var.shape[1])]
vif_df['features'] = var.columns

# checking the results
vif_df
```

Now here is how it works.

When there is a result 1 that means there is no multicolinearity.

Generally the values are between 1 to 5 and are considered okay.
From the results ,it looks like the Year column is too much corelated with other column. and it will down other features values, therefore i will drop the Year column.

```python
# drop the year column
car.drop('Year', axis=1, inplace=True)

car.head()
```

# Now taking care about categorical columns
Brand, Body, Engine Type and Registration

```python
# checking the datatypes 
car.info()
```

```python
car_dum = pd.get_dummies(car, drop_first=True) 
#it will create new column for the categorical values and will increase dimension of the data
# checking the data
car_dum
```

```python
# Checking the Multicollinearity for all columns
# creating a dataframe
vif_df_all = pd.DataFrame()

# adding column
vif_df_all['VIF'] = [vif(car_dum.values, i) for i in range(car_dum.shape[1])]
vif_df_all['features'] = car_dum.columns

# checking the result
vif_df_all
```

all the values are under 5 but the log_price which is our target column has 41.98 as VIF value, this is telling us that the price is highly corellated with other columns
